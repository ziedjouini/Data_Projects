#####visualiser les données et de comprendre la distribution de chaque variable

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Chargez votre DataFrame
# Assumez que votre DataFrame s'appelle 'df'

# Affichez un histogramme pour chaque variable
for column in startups.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(startups[column], kde=True)
    plt.title(f'Distribution de {column}')
    plt.show()
```

```python
# Affichez un boxplot pour chaque variable
for i in startups[['Store ID','Marketing Spend','Revenue']]:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=startups[i])
    plt.title(f'Boxplot de {i}')
    plt.show()
```

```python
# Calculez la matrice de corrélation
corr_matrix = startups.corr()

# Affichez la matrice de corrélation sous forme de heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice de Corrélation')
plt.show()
```

```python
# Affichez un pairplot pour toutes les variables
sns.pairplot(startups)
plt.show()
```
